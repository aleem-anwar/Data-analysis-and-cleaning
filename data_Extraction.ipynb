```python
import pandas as pd 
data = pd.read_csv("Countries.csv")
```

```python
#COUNTRY WITH THE HIGHEST POPULATION
data[data['population'] == data['population'].max()]
```

```python
#CAPITAL OF THE COUNTRY WITH THE HIGHEST POPULATION
data[data['population'] == data['population'].max()]['capital_city']
```

```python
#WHICH COUNTRY HAS THE LEAST POPULATION
data[data['population'] == data['population'].min()]
```

```python
#CAPITAL OF THE COUNTRY WITH LEAST POPULATION
data[data['population'] == data['population'].min()]['capital_city']
```

```python
#TOP 5 COUNTRIES WITH HIGHEST DEMOCRATIC SCORE
data.sort_values(by="democracy_score",ascending=False).head()
```

```python
#total number of regions
tot_reg = data.value_counts("region").sum()
print(f"There are total {tot_reg} regions in this data frame")
```

```python
#number of countries in eastern europe region 
count = data[data['region'] == 'Eastern Europe'].value_counts('country').sum()
print(f"There are total {count} countries in Eastern Europe")
```

```python
#political leader of the country with 2nd highest population
data.nlargest(2,columns='population').iloc[1]['political_leader']
```
