```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
```

```python
x = np.linspace(0,5,10)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(y**2,x) 
plt.subplot(2,2,4) 
plt.plot(x**2,y)
```

```python
canv = plt.figure(dpi = 1800)
a1 = canv.add_axes([0.1,0.1,0.8,0.8])
a1.plot(x,y)
plt.title('X v/s Y graph')
plt.xlabel("X axis")
plt.ylabel("Y axis")
a2 = canv.add_axes([0.25,0.5,0.3,0.3])
a2.plot(x**3,y)
plt.title("X^3 V/S Y graph")
plt.xlabel("X axis") 
plt.ylabel("Y axis")
```
