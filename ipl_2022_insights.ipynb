```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('IPL.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# which team won the most matches 
match_wins = df['match_winner'].value_counts()
sns.barplot(x=match_wins.values, y = match_wins.index,palette='rainbow')
```

```python
#toss decision trends 
df.head()
plt.subplot(1,2,1)
sns.countplot(df['toss_winner'],palette='rainbow')
plt.subplot(1,2,2)
sns.countplot(df['toss_decision'],palette='rainbow',width=0.6)
plt.subplots_adjust(wspace=0.5)
```

```python
#toss winner vs match winner
mt_win = df[df['toss_winner'] == df['match_winner']]['match_id'].count() #number of matches where toss winner was the match winner
print(f"The percentage of matches where toss winner was the match winner is {round((mt_win/df['match_id'].count())*100,2)}%")
```

```python
#won by runs vs wickets 
sns.countplot(df['won_by'],palette='rainbow')
```

```python
# player with the most 'player of the match' awards 
df.head()
plr_of_mt = df['player_of_the_match'].value_counts().head(10)
plr_of_mt
```

```python
sns.barplot(x=plr_of_mt.values,y=plr_of_mt.index,palette='mako')
```

```python
# top 2 scorers 
df.head()
high_scorer = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high_scorer
```

```python
# 10 best bowling figures 
df.head()
def extract_wickets(txt):
    wickets = txt[0] 
    return int(wickets)
def extract_runs(txt): 
    runs = txt[3:] 
    return int(runs) 
df['bbf_wickets'] = df['best_bowling_figure'].apply(extract_wickets) 
df['bf_runs'] = df['best_bowling_figure'].apply(extract_runs)
df.head()
tp_bowlers = df.groupby('best_bowling')['bbf_wickets'].sum().sort_values(ascending=False).head(10)
sns.barplot(x = tp_bowlers.values,y=tp_bowlers.index,palette='mako')
plt.title('TOP 10 MOST WICKET TAKER BOWLERS')
```

```python
#venue where most matches were played 
df.head()
venue_mt_count = df['venue'].value_counts()
sns.barplot(x=venue_mt_count.values,y=venue_mt_count.index,palette='mako')
```
